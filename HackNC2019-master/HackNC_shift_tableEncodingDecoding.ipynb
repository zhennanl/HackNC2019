```python
#shift encoding
import random
alphabet='abcdefghijklmnopqrstuvwxyz'
originalmess=input('Please enter the original message ')
#manual input
#originalmess=''
length=len(originalmess)
key=[0]*length
for i in range(0,length):
    if originalmess[i]==' ':
        key[i]="-"
        continue
    key[i]=random.randint(0,25)
print ('The key for encoding is ')
print (key)
newmess=''
for f in range(0,length):
    char=originalmess[f]
    if char==' ':
        newmess+=' '
        continue
    oriposition=alphabet.find(char)
    newposition=oriposition+key[f]
    if newposition>=26:
        newposition-=26
    newmess+=alphabet[newposition]
print ('The encrypted message is: '+newmess)
```
